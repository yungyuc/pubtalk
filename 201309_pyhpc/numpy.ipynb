Fourier Transform

```python
import numpy as np

from numpy import fft
from matplotlib import pyplot as plt

class Transform(object):
    """Function with Fourier forward transform on a grid.
    """
    def __init__(self, ngrid, extent, average=False):
        self.ngrid = ngrid
        self.extent = extent
        self.interval = interval = extent[1] - extent[0]
        # calculate xgrid.
        self.xgrid = xgrid = np.arange(ngrid, dtype='float64')
        xgrid /= ngrid-1
        xgrid *= interval
        xgrid += extent[0]
        self.dx = dx = xgrid[1] - xgrid[0]
        # calculate bandwidth, kgrid, and kscale.
        self.bw = bw = 1.0 / dx
        self.kgrid = kgrid = np.arange(ngrid, dtype='float64')
        kgrid /= ngrid
        kgrid *= bw
        kgrid -= bw/2
        self.kscale = 1.0 if average else interval/2
        self.kscale /= ngrid/2
        # make x-/k-arrays.
        self.xarrw = np.empty(ngrid, dtype='complex128')
        self.karr = np.empty(ngrid, dtype='complex128')

    def forward(self):
        self.karr[:] = fft.fftshift(fft.fft(self.xarrw))
        self.karr *= self.kscale

    def __str__(self):
        return '; '.join([
            'ngrid: %d' % self.ngrid,
            'extent: %g, %g' % tuple(self.extent),
            'interval: %g' % self.interval,
            'dx: %g' % self.dx,
            'bandwidth: %g' % self.bw,
            'krange: %g, %g' % (self.kgrid[0], self.kgrid[-1])
        ])
```

Fourier Forward Transform for a Sine Function

```python
# Fourier transform for Sine wave.
class SineTransform(Transform):
    def __init__(self, ngrid, extent, freq, **kw):
        super(SineTransform, self).__init__(ngrid, extent, **kw)
        # remember the frequency.
        self.freq = freq
        # initialize x/t data.
        self.xarrw[:] = np.sin(2*np.pi * freq * self.xgrid)
        # for plotting.
        self.fig = None
        self.xax = None
        self.kax = None

    def plot(self, figsize=(12, 6)):
        # create the figure.
        self.fig = fig = plt.figure(figsize=figsize)
        # plot in t/x-space.
        self.xax = xax = fig.add_subplot(1, 2, 1)
        xax.plot(self.xgrid, self.xarrw.real)
        xax.set_title('$N$ = %d' % self.ngrid)
        xax.set_xlim(self.xgrid[0], self.xgrid[-1])
        xax.set_ylim(-1.1, 1.1)
        xax.set_xlabel('$t$/$x$ (s/m)')
        xax.grid()
        # plot in f/k-space.
        self.kax = kax = fig.add_subplot(1, 2, 2)
        kax.plot(self.kgrid, np.absolute(self.karr), label='numpy.fft.fft')
        kax.set_xlim(self.kgrid[0], self.kgrid[-1])
        kax.set_xlabel('$f$/$k$ (Hz/$\\frac{1}{\mathrm{m}}$')
        kax.grid()
        kax.legend()

stfm = SineTransform(2**7, (-1.5, 1.5), 1.0, average=True)
print stfm
stfm.forward()
stfm.plot()
```

Fourier Forward Transform for a Rect Function

```python
# Fourier transform for Rectangle wave.
class RectTransform(Transform):
    def __init__(self, ngrid, extent, **kw):
        super(RectTransform, self).__init__(ngrid, extent, **kw)
        # initialize x/t data.
        self.xarrw.fill(0)
        self.xarrw[np.absolute(self.xgrid) < 0.5] = 1
        self.kana = np.sinc(self.kgrid)
        # for plotting.
        self.fig = None
        self.xax = None
        self.kax = None

    def plot(self, figsize=(12, 6)):
        # create the figure.
        self.fig = fig = plt.figure(figsize=figsize)
        # plot in t/x-space.
        self.xax = xax = fig.add_subplot(1, 2, 1)
        xax.plot(self.xgrid, self.xarrw.real)
        xax.set_title('$N$ = %d' % self.ngrid)
        xax.set_xlim(self.xgrid[0], self.xgrid[-1])
        xax.set_ylim(-0.1, 1.1)
        xax.set_xlabel('$t$/$x$ (s/m)')
        xax.grid()
        # plot in f/k-space.
        self.kax = kax = fig.add_subplot(1, 2, 2)
        kax.plot(self.kgrid, np.absolute(self.karr), label='numpy.fft.fft')
        kax.plot(self.kgrid, np.absolute(self.kana), label='analytical')
        kax.set_xlim(self.kgrid[0], self.kgrid[-1])
        kax.set_xlabel('$f$/$k$ (Hz/$\\frac{1}{\mathrm{m}}$')
        kax.grid()
        kax.legend()

rtfm1 = RectTransform(2**5, (-1., 1.), average=True)
print rtfm1
rtfm1.forward()
rtfm1.plot()
rtfm2 = RectTransform(100, (-5., 5.))
print rtfm2
rtfm2.forward()
rtfm2.plot()
```
