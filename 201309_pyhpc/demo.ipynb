```python
%matplotlib inline # this is a "magic" that allows you to see graphics in the web-based notebook.
from pylab import * # this imports all the functions for plotting.
```

```python
# prepare a sine wave.
xdata = arange(101, dtype='float64')/100*ceil(2*pi)
ydata = sin(xdata)
# plot.
plot(xdata, ydata)
xlabel('$t$ (second)') # show beautiful mathematics in the label.
ylabel('$\sin(t)$')
```
